```python
import numpy as np 
from sklearn import datasets
import pandas as pd 
```

```python
data = datasets.load_iris()

x = data.data 
y = data.target
# initialisation
weight = np.zeros(x.shape[1])
bias = 0
lr = 0.1
epoch = 100 
```

```python
for i in range (epoch) :
    for j in range (len (x)):
        val = np.dot (x[j] , weight) + bias 
        if val >= 0:
            pred = 1 
        else : pred = 0

        # yeh y[j] kanha se laun 
        error = y[j] - pred 
        weight += error * lr * x[j]
        bias += error * lr 

        print (f"weight {i} : {weight}")
```

```python
for i  in range (len (x)):
    val = np.dot (weight , x[i]) + bias 
    if val >= 0 :
        pred = 1 
    else: pred = 0

    print (f"x[{i}] => {pred}")
```
